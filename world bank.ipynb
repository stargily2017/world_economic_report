```python
import pandas as pd
import numpy as np
world_bank = pd.read_excel('C:/Users/starb/Downloads/World+Economic+Indicators\WorldBank.xlsx').rename({"Country Name":"Country"},axis=1)
world_bank.head()
```

```python
world_bank.info()
```

```python
hdi = pd.read_csv('C:/Users/starb/Downloads/World+Economic+Indicators/HDI.csv')
wd_hdi_2014= world_bank.query("Year==2014").merge(hdi[["iso3","hdi_2014"]], how = "left", left_on= "Country Code", right_on= "iso3")




wd_hdi_2014.head()
```

```python
wd_hdi_2014.describe()
```

```python
gdp_pivot = world_bank.pivot_table(index="Year", columns= "Region", values="GDP (USD)", aggfunc="sum")
gdp_pivot.head()
```

```python
pop_pivot = world_bank.pivot_table(index="Year", columns= "Region", values="Population(M)", aggfunc="sum")
pop_pivot.head()
```

```python
wb_hd_region = wd_hdi_2014.groupby("Region").agg({"hdi_2014":"mean"}).sort_values("hdi_2014",ascending=False)
wb_hd_region.head()
```

### Create a stacked area chart showing the growth of GDP over time, with each stack representing a geographic region.

```python
import matplotlib.pyplot as plt
import seaborn as sns
stack_list1= ["#00A3A3","#85D7BB","#8CBEBE","#778899","#CD5C5C","#B22222","#182828"]
fig, ax= plt.subplots()
ax.stackplot(
gdp_pivot.index,
    [gdp_pivot[region] /1_000_000_000_000 for region in gdp_pivot.iloc[-1].sort_values(ascending=False).index],
    
    labels= gdp_pivot.iloc[-1].sort_values(ascending=False).index,
    colors= stack_list1
    )
ax.legend(loc="upper left")
ax.set_title("GDP has grown exponentially over time")
ax.set_ylabel("GDP Trillions")
```

#### How to analyse the graph:
GDP represents a country's total economic output per person. In this case, the GDP graph shows an exponential rise, meaning people are becoming wealthier even as the population grows. This trend highlights the positive impact of industrialization and development on a nations economic well-being. In Python, the iloc[-1] function from the pandas library is often used to access the last row of a DataFrame, which can help analyze the most recent GDP value.


### Create a stacked area chart showing the growth of population over time, with each stack representing a geographic region.

```python
stack_list2= ["#00A3A3","#85D7BB","#8CBEBE","#778899","#CD5C5C","#B22222","#182828"]
fig, ax= plt.subplots()
ax.stackplot(
pop_pivot.index,
    [pop_pivot[region]/1000 for region in pop_pivot.iloc[-1].sort_values(ascending=False).index],
    labels= pop_pivot.iloc[-1].sort_values(ascending=False).index,
    colors= stack_list2
    
        )
ax.legend(loc="upper left")
ax.set_title("Population has surged fron 2B to 7.5B")
ax.set_ylabel("Population Billions")
```

### Create a bubble chart with life expectancy for each country in the x-axis, GDP per capita in the y-axis (log scale), and population as the bubble size. 

```python
import matplotlib.ticker as mticker
bubble_list= ["#00A3A3","#85D7BB","#8CBEBE","#778899","#CD5C5C","#B22222","#182828"]
fig, ax= plt.subplots()

minsize= min(wd_hdi_2014["Population(M)"])
maxsize= max(wd_hdi_2014["Population(M)"])


sns.scatterplot(

data = wd_hdi_2014,
    x="Life expectancy at birth (years)",
    y = "GDP per capita (USD)",
    size= "Population(M)",
    sizes=(minsize,maxsize),
    hue= "Region",
    legend= False,
    palette= bubble_list,
    ax=ax

).set(title= "Life Expectancy Increases as Countries get richer",yscale="log")
formatter = mticker.ScalarFormatter(useOffset=False)
ax.yaxis.set_major_formatter(formatter)
```

### Create a bar chart to show average HDI by region.

```python

import matplotlib.pyplot as plt

# Color list  one per region
bar_list = ["#00A3A3", "#85D7BB", "#8CBEBE", "#778899", "#CD5C5C", "#B22222", "#182828"]

# Create the bar chart directly
plt.bar(wb_hd_region.index, wb_hd_region['hdi_2014'], color=bar_list[:len(wb_hd_region)])

# Add labels and title
plt.title("HDI by Region")
plt.ylabel("Human Development Index (HDI)")
plt.xlabel("Region")

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha="right")

# Display the chart
plt.show()
```

### Create a scatterplot with power consumption in the x-axis and GDP per capita in the y-axis, then use HDI as a color scale for the dots.

```python
sns.scatterplot(
    data=wd_hdi_2014.query("Country != 'Iceland'"),
x="Electric power consumption (kWh per capita)",
y="GDP per capita (USD)",
    hue="hdi_2014",
    palette="coolwarm_r"
    

).set(title="Electricity Drives Development")
```
